```python
#Assign it to a variable called chipo
chipo =pd.read_csv( "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")
print(chipo)
```

```python
#How many products cost more than $10.00?
chipo['item_name'].value_counts()[chipo['item_price'] > 10]
```

```python
#What is the price of each item?
#print a data frame with only two columns item_name and item_price
price = chipo.iloc[:,[2,4]]
print(price)
```

```python
# prompt: Sort by the name of the item

chipo.sort_values('item_name',ascending=True)
```

```python
# prompt: What was the quantity of the most expensive item ordered

most_expensive_item = chipo.loc[chipo['item_price'].idxmax()]
quantity = most_expensive_item['quantity']
print(f"Quantity of the most expensive item ordered: {quantity}")
```
